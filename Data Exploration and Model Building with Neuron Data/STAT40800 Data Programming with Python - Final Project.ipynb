Use a significance level of alpha = 0.01. Display the t-score and p-value for each measurement. Clearly state the conclusion of your tests and explain your reasoning.</a></font>

The t test is a test for the null hypothesis that 2 independent samples have identical average (expected) values. This test assumes that the populations have identical variances by default.

```python
# Loading the required library 

from scipy.stats import ttest_ind
```

```python
# Finding the t-score and p-value for each measurement between the two datasets

cols = ng1_df.columns[1:].tolist()
scores = {}

for column in cols:
    val = ttest_ind(ng1_df[column], ng2_df[column])
    scores[column] = val

pd.DataFrame(scores, index = ['t-score', 'p-value'])
```

Using a significance level of 0.01, from the results above, we can make the following conclusions:

- Since the p-values of all 8 measurements are > 0.01, we fail to reject the null hypothesis (both neuron group 1 and neuron group 2 have identical expected values)
- This tells us that both both neuron group 1 and neuron group 2 are part of the same population having identical expected values and hence there's no difference between the two groups' properties

## <font color = 'green'><a id='the_destination_16'>b) Fit a linear regression model and interpret the fitted model.</a></font>

Used OLS from statsmodels.api to fit a linear regression model where 'total_surface' is the response variable and the remaining 20 variables (including intercept) are the predictor variables. OLS stands for Ordinary Least Squares and is one of the most widely used model due to its efficiency. This model gives best approximate of true population regression line. The principle of OLS is to minimize the square of errors. Based on the results from the summary, we can make the following observations about the model and the variables:

- Based on the t-value, we can see that the features intercept, total_volume, average_diameter and total_length have the highest t-value and hence we can be confident in their co-efficients as predictors for total_surface. Thinking about it intuitively, surface area in general is closely related to diameter, length and volume which is what we're seeing here
- $R^{2}$ is the measurement of how much of the response variable is explained by changes in the predictor variables. $R^{2}$ value of 0.992 tells us that 99.2% variation of 'total_surface' can be explained by the other 20 features. But this is expected due to the high number of features being used as predictor variables
- The adjusted $R^{2}$ penalizes the $R^{2}$ formula based on the number of variables. But even after penalizing, we're getting a 0.991 value which is still very high
- Looking at the p-value and using a significance level of 0.01, we can clearly state that any feature which has a p-value of greater than 0.01 has a high probability of having no effect on the 'total_surface' feature. 13 out of 20 features can be removed when we take a significance level of 0.01
- Multicollinearity is strongly implied by a high condition number, which can be seen in our model. Multicollinearity is a term to describe two or more independent variables that are strongly related to each other and are falsely affecting our predicted variable by redundancy. 
- A Durbin-Watson score of ~2 implies that there's no auto-correlation of the residuals implying that the standard error values are not underestimated and can be trusted

Removal of columns which are correlated with each other along with the non-contributing features will improve the performance and the understanding of the model even further

## <font color = 'green'><a id='the_destination_17'>c) Perform a forward selection Akaike Information Criterion (AIC) regression. Examine the selected model and discuss your findings in relation to the model fitted in part (b).</a></font>

Here we're creating a function to perform forward stepwise selection using AIC/BIC scores keeping AIC score as the default option. For part (c) we're performing AIC regression.

After performing AIC regression, we can see that the model now has 13 predictor variables (including intercept) instead of all 20 (including intercept) as done in part (b). Please find below the observations from this new model:

- Based on the t-value, we continue to see intercept, total_volume, average_diameter and total_length having the highest t-value and the highest confidence as predictors for total_surface
- Despite removing 7 variables from the model, the $R^{2}$ still remains almost same at 0.991 (compared to part (b)) and the adjusted $R^{2}$ is the same as the actual $R^{2}$
- The condition number has drastically reduced from 1.59 x $10^{16}$ to just 60.8 indicating the removal of the multicollinearity problem
- The number of columns with p-value greater than the significance level of 0.01 has now come down to 6 with their p-values showing a considerable reduction
- These 6 columns/features can be described as having no relation to the 'total_surface' feature due to their high p-value (which makes us unable to reject the null hypothesis)

```python
# Creating a function to perform forward stepwise selection. Takes the mode as input (AIC/BIC) with AIC being used as default

import operator
import ast

# Getting all the column names except 'Intercept'
cols = X.columns
cols = cols[1:]

def forwardSelection(X_std, y, mode = 'AIC'):
    
    # Calculating the initial AIC/BIC score with just intercept
    mod = sm.OLS(y, X_std[['intercept']]).fit()
    if mode == 'AIC':
        val = mod.aic
    elif mode == 'BIC':
        val = mod.bic
    
    # Getting the number of features in the predictor variables dataframe
    length = len(X_std.columns)
    
    # Creating an empty dictionary to store the features and their corresponding AIC score
    dict1 = {}
    count = 0
    
    # Adding each feature as the second feature along with 'intercept' to run the first iteration of AIC scores
    # Example - ['intercept', 'average_contraction'], ['intercept', 'average_fragmentation']
    var = []
    for i in cols:
        var.append(['intercept'] + list([i]))
        
    # Running all possible combinations and picking those features which reduce the AIC/BIC score and continuing the process
    while count < (length*(length-1))/2:  # If there are n columns, total relevant combinations possible is n(n-1)/2
        for i in range(len(var)):
            count+=1
            mod = sm.OLS(y, X_std[var[i]]).fit()
            if mode == 'AIC' and mod.aic < val:
                val = mod.aic
                dict1[str(var[i])] = val
            elif mode == 'BIC' and mod.bic < val:
                val = mod.bic
                dict1[str(var[i])] = val
            else:
                continue  
        
        # Updating the val variable with lowest AIC/BIC value obtained which will be stored as the last value in the dictionary
        val = list(dict1.values())[-1]
        
        # Getting the corresponding feature names with the lowest AIC score
        col_names = list(dict1.keys())[-1]
        
        # Creating the next set of possible feature names which can be used to calc the AIC/BIC scores and repeating the process
        var = []
        col_names = ast.literal_eval(col_names)
        for i in cols:
            if i in col_names:
                continue
            else:
                var.append(col_names + [i])
    return dict1.items()
```

```python
# Storing the result as a dataframe and showing the final set of features selected based on forward stepwise selection using AIC

val1 = pd.DataFrame(forwardSelection(X_std, y, 'AIC'), columns = ['Features', 'Score'])
val1['Features'].values[-1]
```

```python
# Using the final set of selected features and fitting a linear regression model

mod = sm.OLS(y, X_std[ast.literal_eval(val1['Features'].values[-1])]).fit()
print(mod.summary())
```

## <font color = 'green'><a id='the_destination_19'>e) Explain how using BIC for model selection differs from using AIC.</a></font>

AIC and BIC are widely used in model selection criteria. AIC means Akaike’s Information Criterion and BIC means Bayesian Information Criterion. Though these two terms address model selection, they are not the same. Please find below some of the differences between the 2 criterias:


- AIC chooses the threshold according to how many degrees of freedom the variable under consideration has. The more degrees of freedom a variable has, the lower the threshold will be.
- BIC chooses the threshold according to the effective sample size n. The larger n is, the lower the threshold will be.
- The AIC can be termed as a measure of the goodness of fit of any estimated statistical model. The BIC is a type of model selection among a class of parametric models with different numbers of parameters
- Penalty for additional parameters is more in BIC than AIC. Unlike the AIC, the BIC penalizes free parameters more strongly
- BIC is a more restrictive criterion than AIC and so yields smaller models (as was illustrated in our model in section (d))

## <font color = 'green'><a id='the_destination_24'>d) Assess the performance of a random forest regression model with 5, 10, 20, 50, 100, 200, 500 and 1000 trees in predicting the total surface area of a neuron. You should repeat the model fit and prediction 30 times for each number of trees, using a different random state for each repeat. Create a plot of the model performance as a function of the number of trees (use a log axis for the number of trees). The plot should show the mean and standard error of the performance metric for each number of trees. Discuss your findings.</a></font>

Created a list containing the 8 different tree sizes and 30 different random states and performed the random forest regression on the training data and used the test data and predicted data to assess the model performance by calculating ghe Mean Squared Error. Got the Mean and the Standard Error (Standard deviation/sqrt(Number of samples)) of the 30 MSE calculated for each random state, for each tree size and then used matplotlib to plot the mean and standard error against the log of number of trees.

Looking at the plot, we can understand the following:
- The Mean and the Standard Error of the MSE (performance metric) comes down as the number of trees increases
- Lower the number of trees, higher the standard deviation of the MSE 
- The level of reduction of the Mean and Standard Error comes down as we reach higher number of trees 
    - Standard Error reduced by ~10K when the trees increased from 5 to 10, but reduced by only another ~20K when the trees increased from 10 to 500

```python
# Ran the random forest regression for different tree sizes and multiple random states to assess model performance

from sklearn.metrics import mean_squared_error

trees = [5, 10, 20, 50, 100, 200, 500, 1000]
rs = [i for i in range(1,31)]

mse_mean_dict = {}
mse_se_dict = {}
mse_list = []

for tree in trees:
    for val in rs:
        rf = RandomForestRegressor(n_estimators = tree, random_state = val)
        rf.fit(X_train, y_train)
        rf_test_pred  = rf.predict(X_test)
        mse = mean_squared_error(y_test, rf_test_pred)
        mse_list.append(mse)
        
    mse_arr = np.array(mse_list)
    mse_mean = np.mean(mse_arr)
    mse_std = np.std(mse_arr)
    mse_se = mse_std/np.sqrt(len(rs))
    mse_mean_dict[str(tree)] = mse_mean
    mse_se_dict[str(tree)] = mse_se
```

Created a new dataframe containing only the morphological features (excluded 'id') and ran the clustering algorithm with different cluster sizes (1-10). Then plotted the SSE vs the number of clusters to get the optimal number of clusters to be used. 
The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid. The way we can find the optimal number of clusters is by using the Elbow Criterion. The idea of the Elbow Criterion method is to choose the k(no of clusters) at which the SSE decreases abruptly. SSE tends to decrease toward 0 as we increase k (SSE=0, when k is equal to the no. of data points in the dataset, because then each data point is its own cluster, and there is no error between it and the center of its cluster). In our case, when k = 4 seems to satisfy this criteria. Hence we'll proceed with k = 4.

```python
# Creating a new dataframe that contains all the morphological features (except id)

df_cluster = df.drop('id', axis = 1)
df_cluster.head()
```

```python
# Plotting the SSE vs Number of clusters (k) curve

from sklearn.cluster import KMeans

clusters = [i for i in range(1,10)]

sse = {}
for cluster_count in clusters:
    model = KMeans(n_clusters = cluster_count, random_state = 42)
    model.fit(df_cluster)
    sse[cluster_count] = model.inertia_
    
plt.figure(figsize=(8,8))
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel('Number of Clusters (k)', fontsize = 12)
plt.ylabel('SSE', fontsize = 12)
plt.title('Elbow Criterion')
plt.show()
```

<b>Agglomerative clustering</b>

The Agglomerative clustering is the most common type of hierarchical clustering used to group objects in clusters based on their similarity. It’s also known as AGNES (Agglomerative Nesting). The algorithm starts by treating each object as a singleton cluster. Next, pairs of clusters are successively merged until all clusters have been merged into one big cluster containing all objects.

Agglomerative clustering works in a “bottom-up” manner. That is, each object is initially considered as a single-element cluster (leaf). At each step of the algorithm, the two clusters that are the most similar are combined into a new bigger cluster (nodes). This procedure is iterated until all points are member of just one single big cluster (root) (see figure below).

Typical steps in the process:
- Each data point is treated as a single cluster. We have K clusters in the beginning. At the start, the number of data points will also be K.
- Now we need to form a big cluster by joining 2 closest data points in this step. This will lead to total K-1 clusters.
- Two closest clusters need to be joined now to form more clusters. This will result in K-2 clusters in total. 
- Repeat the above three steps until K becomes 0 to form one big cluster. No more data points are left to join.
- After forming one big cluster at last, we can use dendrograms to split the clusters into multiple clusters depending on the use case.


<b>Agglomerative clustering Use Cases</b>

Agglomerative clustering has its application in many areas which includes: Segmentation, Data Mining, Entity Resolution, Speech Separation.
